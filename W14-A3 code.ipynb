```python
def func(**kwargs):
    print(kwargs)

#**kwargs passes a dictionary of arguments to the function

func(a=1, b=2)

data = {'a': 1, 'b': 2}
func(**data)
```

```python
def func(*args):
    print(args)

#*args passes a tuple of arguments to the function

func(1, 2, 3)

nums = [1, 2, 3]
func(*nums)
```
